### Visualization

```python
import pandas as pd
```

```python
data=pd.read_csv('titanic.csv',index_col=0)
```

```python
data.columns
```

```python
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.countplot(x='Age',data=df)
```

```python
sns.countplot(x='Sex',data=df)
```

```python
sns.countplot(x='SibSp',data=df)
```

```python
sns.countplot(x='Pclass',data=df)
```

```python
sns.countplot(x='Survived',data=df)
```

### stat in ratio

```python
sns.countplot(x='Sex',hue='Survived',data=df)
```

```python
sns.countplot(x='Parch',hue='Survived',data=df)
```

```python

sns.relplot(x='Age',y='Fare',hue='Survived',data=df)
```

```python
sns.catplot(x='Sex',y='Fare',hue='Survived',data=df)
```

```python
sns.catplot(x='Sex',y='Age',hue='Survived',data=df)
```

```python
sns.catplot(x='Pclass',y='Age',hue='Survived',data=df)
```

```python
sns.catplot(x='Fare',y='Pclass',hue='Survived',data=df)
```
